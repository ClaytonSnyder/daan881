```python
import pandas as pd
```

# MITRE Dataset: Common Vulnerabilities and Exposures (CVE) 

https://cve.mitre.org/data/downloads/index.html  

```python
cve_master_list = pd.read_csv('data/allitems.csv', names=["cve","Status","Description","References","Phase","Votes","Comments"])

print(cve_master_list.shape)
cve_master_list.head()
```

# EPSS Dataset

https://www.first.org/epss/data_stats

```python
epss_list = pd.read_csv('data/epss_scores-2024-09-17.csv')

print(epss_list.shape)
epss_list.head()
```

# CISA Known Exploited Vulnerabilities

https://github.com/jgamblin/CISA_Enrichment?tab=readme-ov-file

```python
cisa_enriched = pd.read_csv('data/cisa_enriched.csv')
cisa_enriched.set_index('CVE')
cisa_enriched = cisa_enriched.drop(['Vulnerability_Name', 'Description', 'Vendor', 'Product', 
                                    'Required_Action', 'CVE_Published', 'CVE_Modified', 'References', 'Vulnerable_Products'], axis=1)

print(cisa_enriched.shape)
cisa_enriched.head()
```

# Big Vul Dataset

https://github.com/ZeoVan/MSR_20_Code_vulnerability_CSV_Dataset?tab=readme-ov-file

```python
big_vul = pd.read_csv('data/MSR_data_cleaned.csv')
big_vul = big_vul[['Access Gained', 'Attack Origin',
       'Authentication Required', 'Availability', 'CVE ID',
       'CWE ID', 'Complexity', 'Confidentiality', 'Integrity',
       'Known Exploits', 'Publish Date', 'Score', 'Update Date',
       'Vulnerability Classification']]
print(big_vul.shape)

big_vul.head()
```

```python
big_vul = big_vul.drop_duplicates(subset='CVE ID', keep='first')

print(big_vul.shape)

big_vul.head()
```

```python
combined = pd.merge(epss_list, cve_master_list, on='cve')
combined = pd.merge(combined, cisa_enriched, left_on='cve', right_on='CVE', how='left')
combined = pd.merge(combined, big_vul, left_on='cve', right_on='CVE ID', how='left')
combined = combined.drop(['CVE', 'CVE ID'], axis=1)
combined = combined.drop_duplicates()
print(combined.shape)
combined.head()
```

```python
combined.to_csv('data/DAAN881_group4_dataset.csv', index=False)  
```
